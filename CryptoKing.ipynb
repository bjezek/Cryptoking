```python
# look for arbitrage later in the month to see opportunity for late in the month subtract bitstamp march from coinbase march and calculate spread
#for

arbitrage_spread_late = bitstamp_sliced['Close'].loc['2018-03-26'] - coinbase_sliced['Close'].loc['2018-03-26']
arbitrage_spread_late.describe()
```

```python
#late opportunites for late day returns greater than 0 divided by the close price of March 26th 

late_spread_return = arbitrage_spread_late[arbitrage_spread_late>0]/bitstamp['Close'].loc['2018-03-26']
late_spread_return
```

```python
# profit of spread opportunity greater than 1cent 

profit_late_day = late_spread_return[late_spread_return> 0.01]
profit_late_day
```

```python
#Calculate profit total late day sum 

profit_late_day.sum()
```

```python
#Calculate late day profit for daily gain

cumulative_late_day = profit_late_day.cumsum()
cumulative_late_day
```

```python
#graph late day profits showing opportunites 

cumulative_late_day.plot(figsize=(10,7), title="Cumulative Late Profits")
```

```python
#Close spread late for plot box graph

arbitrage_spread_late.plot(kind='box')
```

```python
# middle spread arbitrage opportunites in Febuary 1st to the 24th for profit opportunites.

middle_spread_return = arbitrage_spread_middle[arbitrage_spread_middle>0]/bitstamp['Close'].loc['2018-02-01':'2018-02-24']
middle_spread_return.dropna()
```

```python
#spread for middle of month thats greater than 0.01 through the middle of the month per day

profit_middle_day = profit_trade_middle = middle_spread_return[middle_spread_return> 0.01]
profit_middle_day
```

```python
# profit per day in the middle of the moth 

profit_middle = profit_middle_day.sum()
profit_middle
```

```python
# total of cumulative profit for mid-month

cumulative_middle_profit = profit_middle_day.cumsum()
cumulative_middle_profit
```

```python
#plot cumulative profit for middle time frame of data

cumulative_middle_profit.plot(figsize=(10,7), title="Cumulative Profit Profits")
```
